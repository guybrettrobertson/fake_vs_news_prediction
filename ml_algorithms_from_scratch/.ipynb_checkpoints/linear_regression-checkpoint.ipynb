# Linear regression

```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Linear regression class object

```python
class LinearRegression(object):
    def __init__(self):
        """Class constructor."""
        
    def fit(self, x, y):
        """Fit a straight line to the data."""
        
        self.x = x
        self.y = y
        
        x = np.c_[np.ones((len(y), 1)), x]
        self.params = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)

    def predict(self, x):
        """Predicts a y for a given x."""
        
        return self.params[0] + self.params[1] * x
```

## Testing

```python
np.random.seed(42)
x = 2 * np.random.rand(1000, 1)
y = 4 + 3 * x + np.random.randn(1000, 1)
```

```python
line = LinearRegression()
```

```python
line.fit(x, y)
```

```python
line.params
```

```python
line.predict([0, 0])
```

## Visualisation

```python
x = x.reshape((len(x),))
y = y.reshape((len(x),))
input_data = pd.DataFrame(data = {'x': x, 'y': y}, columns = ['x', 'y'])

x_model = np.array([0, 2])
y_model = line.predict(x_model)
model_data = pd.DataFrame(data = {'x_model': x_model, 'y_model': y_model}, columns = ['x', 'y'])
```

```python
sns.scatterplot(x, y, data = input_data, alpha = 0.3)
sns.lineplot(x_model, y_model, data = model_data, color = 'red', size = 100, legend = False)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
