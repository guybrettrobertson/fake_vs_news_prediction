# Data preprocessing

## Import packages

```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from gensim.corpora.dictionary import Dictionary
from collections import defaultdict
```

## Read in data

```python
complaints_df = pd.read_pickle('~/documents/data/consumer_complaints/consumer_complaints_pre-processed.csv')
n_rows = complaints_df.shape[0]
```

## Term frequency - inverse document frequency

Create a dictionary of all the words in the complaints data.

```python
all_complaints = complaints_df['Consumer complaint narrative'].values.tolist()

dictionary = Dictionary(all_complaints)
```

```python
dictionary.token2id
```

Create a corpus for the complaints.

```python
corpus = [dictionary.doc2bow(complaint) for complaint in all_complaints]
```

```python
print(corpus[0:2])
```

```python
complaint = corpus[0]
```

Sort the complaint for frequency to produce a bag of words.

```python
bow_complaint = sorted(complaint, key=lambda w: w[1], reverse = True)
```

Top 5 words.

```python
for word_id, word_count in bow_complaint[:5]:
    print(dictionary.get(word_id), word_count)
```
